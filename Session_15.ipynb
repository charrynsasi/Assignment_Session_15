Problem Statement 1:
You survey households in your area to find the average rent they are paying. Find the
standard deviation from the following data:
$1550, $1700, $900, $850, $1000, $950.

```python
import numpy as np
lst=[1550,1700,900,850,1000,950]
print('The standard deviation is ${:.2f}'.format(np.std(lst)))
```

Problem Statement 2:
Find the variance for the following set of data representing trees in California (heights in
feet):
3, 21, 98, 203, 17, 9

```python
lst=[3, 21, 98, 203, 17, 9]
print('The variance is {:.2f}'.format(np.var(lst)))
```

Problem Statement 3:
In a class on 100 students, 80 students passed in all subjects, 10 failed in one subject, 7
failed in two subjects and 3 failed in three subjects. Find the probability distribution of
the variable for number of subjects a student from the given class has failed in.

```python
#The probability of failing in 0 subjects, P(X=0) = 80/100
#The probability of failing in 1 subjects, P(X=1) = 10/100
#The probability of failing in 2 subjects, P(X=2) = 7/100
#The probability of failing in 3 subjects, P(X=3) = 3/100

#The probability distribution P(X) of the variable X for number of subjects a student from the given class has failed in is :

# X     0    1    2     3
# P(X)  0.8  0.1  0.07  0.03



%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
x=[0,1,2,3]
y=[0.8,0.1,0.07,0.03]

fig = plt.figure()
ax = plt.axes()
ax.plot(x,y);
plt.title('The probability distribution')
plt.xlabel("Variable X")
plt.ylabel('The probability P(X)');
```
